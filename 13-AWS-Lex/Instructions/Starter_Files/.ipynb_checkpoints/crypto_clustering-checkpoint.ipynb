# Clustering Crypto

```python
# Initial imports
import requests
import pandas as pd
import matplotlib.pyplot as plt
import hvplot.pandas
import plotly.express as px
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
# Alternatively, use the provided csv file:
# file_path = Path("Resources/crypto_data.csv")
from pathlib import Path
```

```python
# Create a DataFrame
file_path = Path("Resources/crypto_data.csv")
coin_df = pd.read_csv(file_path)
coin_df.head()
```

### Data Preprocessing

```python
# Keep only necessary columns:
# 'CoinName','Algorithm','IsTrading','ProofType','TotalCoinsMined','TotalCoinSupply'
coin_df = coin_df[['CoinName','Algorithm','IsTrading','ProofType','TotalCoinsMined','TotalCoinSupply']]
coin_df.head()
```

```python
# Keep only cryptocurrencies that are trading
coin_df = coin_df[coin_df.IsTrading != False]
coin_df
```

```python
# Keep only cryptocurrencies with a working algorithm
# Keep only cryptocurrencies that are trading
coin_df = coin_df.dropna(subset = ['Algorithm'])
coin_df = coin_df[coin_df.Algorithm != 'N/A']
coin_df
```

```python
# Remove the "IsTrading" column
coin_df = coin_df.drop(columns = ['IsTrading'])
coin_df
```

```python
# Remove rows with at least 1 null value
coin_df = coin_df.dropna()
coin_df
```

```python
# Remove rows with cryptocurrencies having no coins mined
coin_df = coin_df[coin_df.TotalCoinsMined > 0]
coin_df
```

```python
# Drop rows where there are 'N/A' text values
coin_df = coin_df.dropna()
coin_df
```

```python
# Store the 'CoinName'column in its own DataFrame prior to dropping it from crypto_df
coinname_df = pd.DataFrame()
coinname_df['CoinName'] = coin_df['CoinName']
coinname_df
```

```python
# Drop the 'CoinName' column since it's not going to be used on the clustering algorithm
coin_df = coin_df.drop(columns = ['CoinName'])
coin_df
```

```python
# Create dummy variables for text features
X = pd.get_dummies(crypto_df, columns = ["Algorithm","ProofType"],drop_first = True)
X.head()
```

```python
# Standardize data
from sklearn.preprocessing import StandardScaler
```

```python
coin_scaled = StandardScaler().fit_transform(X)
print(coin_scaled)
```

### Reducing Dimensions Using PCA

```python
from sklearn.decomposition import PCA
```

```python
# Use PCA to reduce dimensions to 3 principal components
pca = PCA(n_components=3)

coin_pca = pca.fit_transform(coin_scaled)
```

```python
# Create a DataFrame with the principal components data
coin_pca = pd.DataFrame(
    data=coin_pca, columns=["principal component 1", "principal component 2", "principal component 3"]
)
coin_pca.head()
```

### Clustering Crytocurrencies Using K-Means

#### Find the Best Value for `k` Using the Elbow Curve

```python
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of k values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(coin_pca)
    inertia.append(km.inertia_)

# Create the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", title="Elbow Curve", xticks=k)
```

Running K-Means with `k=<your best value for k here>`

```python
# Initialize the K-Means model
model = KMeans(n_clusters = 4)

# Fit the model
model.fit(coin_pca)

# Predict clusters
predictions = model.predict(coin_pca)

# Create a new DataFrame including predicted clusters and cryptocurrencies features
coin_pca["class"] = model.labels_
coin_pca.head()
```

### Visualizing Results

#### 3D-Scatter with Clusters

```python
# Create a 3D-Scatter with the PCA data and the clusters
fig = px.scatter_3d(
    coin_pca,
    x="principal component 1",
    y="principal component 2",
    z="principal component 3",
    color="class",
    symbol="class",
    width=800,
)
fig.update_layout(legend=dict(x=0, y=1))
fig.show()
```

#### Table of Tradable Cryptocurrencies

```python
# Table with tradable cryptos
merged_coins = pd.concat([coinname_df,coin_df,coin_pca], axis = 1, sort = False)
coin_table = merged_coins.hvplot.table()
coin_table
```
